```julia
using Plots
```

```julia
# y = sint ^2 
f(x) = (sin(x))^2
plot(f, -4pi, 4pi, framestyle=:origin)
s(x) = 1/2 - 1/2 * cos(2*x)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = cost ^2
f(x) = (cos(x))^2
plot(f, -4pi, 4pi, framestyle = :origin)
s(x) = 1/2 + 1/2 * cos(2*x)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = sin^2(t)* cos^2(t)
f(t) = (sin(t)*cos(t))^2
plot(f, -4pi, 4pi, framestyle=:origin)
s(t) = 1/8 - 1/8 * cos(4 * t)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = cos^4(t)
f(t) = (cos(t))^4
plot(f, -pi, pi, framestyle=:origin)
s(t) = 3/8 + 1/2 *cos(2*t) + 1/8 * cos(4*t)
plot!(s)
```

```julia
# zad 2 a)
function f(t)
   if floor(t/pi) % 2 == 0
        return 0
    else
        return 1
    end
end
plot(f, -6, 6, framestyle=:origin)
s(t) = 1/2 + sum((cos(i*pi)-1)*sin(i*t)/(i*pi) for i in 1:100000)
plot!(s, -6, 6)
```

```julia
# zad 2 b)
function f(t)
   if floor(t/pi) % 2 == 0
        return t - pi*floor(t/pi)
    else
        return 1
    end
end

plot(f, -12, 12, framestyle=:origin)
s(t) = (1 + pi/2)/2 + sum((cos(i*pi)-1)*cos(i*t)/(pi*i^2) + sin(i*t)*(cos(i*pi)-1-pi*cos(i*pi))/(i*pi) for i in 1:10000)
plot!(s, -12, 12)
```

```julia
# zad 2 C)
function f(t)
   if floor(t/pi) % 2 == 0
        return pi - t + pi*floor(t/pi)
    else
        return t - pi*floor(t/pi) - pi
    end
end

plot(f, -12, 12, framestyle=:origin)
s(t) = sum((1-cos(i*pi))*(2*cos(i*t)/i/pi + sin(i*t))/i for i in 1:1000)
plot!(s)
```

```julia
# zad 2 d)
f(t) = abs(sin(t))
plot(f, -12, 12, framestyle=:origin)
s(t) = 2/pi + sum(-cos(i*t)/pi * ((cos(pi+i*pi)-1)/(i+1)+(cos(pi-i*pi)-1)/(1-i)) for i in 2:100)
plot!(s, marker=true)
```

```julia
# zad 2 e)
# f(t) = (t - pi*floor(t/pi))^2
function f(t)
    if floor(t/pi) % 2 == 0
        return (t - pi*floor(t/pi))^2
    else
        return (t - pi*floor(t/pi) - pi)^2
    end
end
plot(f, -12, 12, framestyle=:origin)
s(t) = pi^2/3 + sum(4*cos(i*pi)*cos(i*t)/i^2 for i in 1:100)
plot!(s)
```
