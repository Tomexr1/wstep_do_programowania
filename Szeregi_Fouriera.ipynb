```julia
using Plots
```

```julia
# y = sint ^2 
f(x) = (sin(x))^2
plot(f, -4pi, 4pi, framestyle=:origin)
s(x) = 1/2 - 1/2 * cos(2*x)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = cost ^2
f(x) = (cos(x))^2
plot(f, -4pi, 4pi, framestyle = :origin)
s(x) = 1/2 + 1/2 * cos(2*x)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = sin^2(t)* cos^2(t)
f(t) = (sin(t)*cos(t))^2
plot(f, -4pi, 4pi, framestyle=:origin)
s(t) = 1/8 - 1/8 * cos(4 * t)
plot!(s, -4pi, 4pi, framestyle=:origin)
```

```julia
# y = cos^4(t)
f(t) = (cos(t))^4
plot(f, -pi, pi, framestyle=:origin)
s(t) = 3/8 + 1/2 *cos(2*t) + 1/8 * cos(4*t)
plot!(s)
```
